# Combined State

```python
df0 = pd.read_csv('./data/data_cleaned/combined_data_state_level_na_filled_from_2000.csv')
```

```python
df0.head()
```

```python
df0.columns = df0.columns.str.lower()
```

```python
df0.isnull().sum()
```

```python
df0 = df0.drop('edu_expenditure_percent_gov_income', axis = 1)
```

```python
df0.columns
```

```python
df0.head()
```

```python
to_sum_feature = ['year', 'population', 'known_property_crimes', 'teachers_elem_secd_schools',  
                'pub_high_school_graduates', 
                 ]
to_avg_feature = ['year', 'avg_attendence_elem_secd_schools','pupil_teacher_ratio','high_school_completers_18_to_24',
                  'bachelor_completers_over_25', 'expenditure_per_pupil', 'gdp_per_capita', 'income_per_capita', 
                  'unemployment_rate','expenditure_per_pupil']
```

```python
df1 = df0[to_sum_feature].groupby('year').sum().reset_index()
df1.head()
```

```python
df2 = df0[to_avg_feature].groupby('year').mean().reset_index()
df2.head()
```

```python
df0 = df1.join(df2.set_index('year'), how = 'left', on = 'year')
df0.head()
```

```python
df = df.join(df0.set_index('year'), how = 'left', on = 'year')
```

```python
df.head()
```

# US dollar index

```python
df0 = pd.read_csv('./data/data_raw/us-dollar-index-historical-chart.csv', skiprows=14)
```

```python
df0.columns = ['date', 'dollar_index']
df0.date = pd.to_datetime(df0.date)
```

```python
df0.head()
```

```python
df0 = df0.set_index('date')
df0 = df0.resample('Y').mean().reset_index()
```

```python
df0.head()
```

```python
df0.date= df0.date.apply(lambda x: int(str(x)[:4]))
```

```python
df0.head()
```

```python
df = df.join(df0.set_index('date'), how = 'left', on='year')
```
