# Combined State

```python
df0 = pd.read_csv('./data/data_cleaned/combined_data_state_level_na_filled.csv')
```

```python
df0.head()
```

```python
df0.columns = df0.columns.str.lower()
```

```python
to_sum_feature = ['year','population', 'known_property_crimes']
to_avg_feature = ['year', 'expenditure_per_pupil', 'gdp_per_capita', 'income_per_capita', 'unemployment_rate']
```

```python
df1 = df0[to_sum_feature].groupby('year').sum().reset_index()
df1.head()
```

```python
df2 = df0[to_avg_feature].groupby('year').sum().reset_index()
df2.head()
```

```python
df0 = df1.join(df2.set_index('year'), how = 'left', on = 'year')
df0.head()
```

```python
df = df.join(df0.set_index('year'), how = 'left', on = 'year')
```

# US dollar index

```python
df0 = pd.read_csv('./data/data_raw/us-dollar-index-historical-chart.csv', skiprows=14)
```

```python
df0.columns = ['date', 'dollar_index']
df0.date = pd.to_datetime(df0.date)
```

```python
df0.head()
```

```python
df0 = df0.set_index('date')
df0 = df0.resample('Y').mean().reset_index()
```

```python
df0.head()
```

```python
df0.date= df0.date.apply(lambda x: int(str(x)[:4]))
```

```python
df0.head()
```

```python
df = df.join(df0.set_index('date'), how = 'left', on='year')
```
